```python
#### mettre toute le columns en minuscule et convertir en chaine de caractère
for df5 in [carac , lieu , usag , vehi] :
    df5.rename(columns =str.lower, inplace=True)
    df5 ["num_acc"] = df5["num_acc"].astype(str)
if "num_veh" in vehi.columns:
    vehi["num_veh"] = vehi["num_veh"].astype(str)
if "num_veh" in usag.columns: 
    vehi["num_vehi"]  = vehi["num_veh"].astype(str)
```

```python
df5 = carac.merge(lieu , on ="num_acc" , how="left" , suffixes=("", "_lieu"))
df5 = df5.merge(vehi , on ="num_acc" , how="left" , suffixes=("", "_vehi"))
df5 = df5.merge(usag , on =["num_acc" ,"num_veh"] , how="left" , suffixes=("", "_usag"))
```
